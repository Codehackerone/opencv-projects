```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread('../DATA/sammy_face.jpg')
plt.imshow(img)
```

```python
edges=cv2.Canny(image=img,threshold1=127,threshold2=127)
plt.imshow(edges)
```

```python
med_val=np.median(img) 
```

```python
med_val
```

```python
# LOWER THRESHOLD TO EITHER 0 or 75% OF MEDIAN VALUE OF WHICHEVER IS GREATER
lower=int(max(0,0.7*med_val))
# LOWER THRESHOLD TO EITHER 130% OF MEDIAN VALUE OR MAX 255 OF WHICHEVER IS GREATER
upper=int(max(1.3*med_val,255))
```

```python
edges=cv2.Canny(image=img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img=cv2.blur(img,ksize=(5,5))
```

```python
edges=cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```
